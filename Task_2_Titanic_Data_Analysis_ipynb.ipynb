```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
df = pd.read_csv('train.csv')
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
# Fill missing Age values with median age
df['Age'].fillna(df['Age'].median(), inplace=True)

# Fill Embarked with mode
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)

# Drop Cabin because too many missing values
df.drop('Cabin', axis=1, inplace=True)

# Verify
df.isnull().sum()
```

```python
survival_gender = df.groupby('Sex')['Survived'].mean()
print(survival_gender)
```

```python
plt.figure(figsize=(6,4))
sns.barplot(x='Sex', y='Survived', data=df)
plt.title('Survival Rate by Gender')
plt.show()
```

```python
class_survival = df.groupby('Pclass')['Survived'].mean()
print(class_survival)
```

```python
plt.figure(figsize=(6,4))
sns.barplot(x='Pclass', y='Survived', data=df)
plt.title('Survival Rate by Passenger Class')
plt.show()
```

```python
df['AgeGroup'] = pd.cut(
    df['Age'],
    bins=[0,12,18,35,60,100],
    labels=['Child','Teen','Young Adult','Adult','Senior']
)
```

```python
age_survival = df.groupby('AgeGroup')['Survived'].mean()
print(age_survival)
```

```python
plt.figure(figsize=(8,5))
sns.barplot(x='AgeGroup', y='Survived', data=df)
plt.title('Survival Rate by Age Group')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df['Age'], bins=20, kde=True)
plt.title('Distribution of Passenger Ages')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x='Pclass', hue='Survived', data=df)
plt.title('Passenger Class vs Survival')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.countplot(x='Sex', hue='Survived', data=df)
plt.title('Gender vs Survival')
plt.show()
```

Conclusion

1. Female passengers survived more than male passengers.
2. Passenger class strongly affected survival chances.
3. First-class passengers had the highest survival rate.
4. Children generally had better survival rates.
5. Most passengers were between 20 and 40 years old.

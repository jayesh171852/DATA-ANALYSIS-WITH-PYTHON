```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Configure visualization style
sns.set_theme(style='whitegrid')
%matplotlib inline
```

```python
import os
from google.colab import files

# Option to manually upload the dataset
if not os.path.exists('student-mat.csv'):
    print("Please upload 'student-mat.csv' below:")
    uploaded = files.upload()

# Fallback: Check if the dataset exists after upload, if not, download it
if not os.path.exists('student-mat.csv'):
    print("File not uploaded. Downloading student-mat.csv as fallback...")
    import urllib.request
    url = "https://raw.githubusercontent.com/yofisunarto/student-performance-dataset/master/student-mat.csv"
    urllib.request.urlretrieve(url, 'student-mat.csv')

# Load the dataset using semicolon as the separator
df = pd.read_csv('student-mat.csv', sep=';')

# Verify successful loading
print("\nDataset loaded successfully.")
print(f"Total records: {len(df)}")
```

```python
# Display the first 5 rows
print("--- First 5 Rows ---")
display(df.head())

# Display dataset shape (rows, columns)
print(f"\nDataset Shape: {df.shape}")

# Display column names
print(f"\nColumn Names: {df.columns.tolist()}")

# Display data types of each column
print("\n--- Data Types ---")
print(df.dtypes)

# Display statistical summary for numerical columns
print("\n--- Statistical Summary ---")
display(df.describe())
```

```python
# Check for missing values in each column
print("Missing values per column:")
print(df.isnull().sum())

# Check for duplicate rows
duplicates = df.duplicated().sum()
print(f"\nNumber of duplicate rows: {duplicates}")

# Remove duplicates if any exist
if duplicates > 0:
    df = df.drop_duplicates()
    print("Duplicates removed.")
```

```python
# 1. Calculate the average final grade (G3)
avg_g3 = df['G3'].mean()
print(f"Average Final Grade (G3): {avg_g3:.2f}")

# 2. Count students who scored above 15 in G3 (High Performers)
high_scorers = df[df['G3'] > 15].shape[0]
print(f"Students scoring above 15 in G3: {high_scorers}")

# 3. Calculate correlation between studytime and G3
correlation = df['studytime'].corr(df['G3'])
print(f"Correlation between Study Time and G3: {correlation:.4f}")

# 4. Compare average G3 grades between male (M) and female (F) students
gender_comparison = df.groupby('sex')['G3'].mean()
print("\nAverage G3 by Gender:")
print(gender_comparison)
```

```python
# Initialize the figure for multiple plots
plt.figure(figsize=(18, 5))

# 1. Histogram of final grades (G3)
plt.subplot(1, 3, 1)
sns.histplot(df['G3'], bins=15, kde=True, color='skyblue')
plt.title('Distribution of Final Grades (G3)')

# 2. Scatter plot of studytime vs G3
plt.subplot(1, 3, 2)
sns.scatterplot(x='studytime', y='G3', data=df, alpha=0.6)
plt.title('Study Time vs Final Grade')

# 3. Bar chart comparing average grades by gender
plt.subplot(1, 3, 3)
sns.barplot(x=gender_comparison.index, y=gender_comparison.values, palette='pastel')
plt.title('Average Grade by Gender')
plt.ylabel('Average G3')

plt.tight_layout()
plt.show()
```
